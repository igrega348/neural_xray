```python
import yaml
import numpy as np
from pathlib import Path
import torch
import matplotlib.pyplot as plt
from objects import Object
```

```python
lat = Object.from_yaml(Path('./lattice.yaml'))
```

```python
x = torch.linspace(-1,1,200)
y = torch.linspace(-1,1,200)
X,Y = torch.meshgrid(x,y, indexing='ij')
pos = torch.stack([X.flatten(),Y.flatten(),X.flatten()*0],dim=1)
rho = lat.density(pos).view(X.shape)
```

```python
plt.imshow(rho)
plt.show()
```
